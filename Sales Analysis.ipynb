```python
sales_df['year_month'] = sales_df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
sales_df['year_month'] 
```

```python
sales_df.groupby('year_month').sum()
```

```python
sales_trend = sales_df.groupby('year_month').sum()['sales'].reset_index()
```

```python
sales_trend
```

```python
plt.plot(sales_trend['year_month'], sales_trend['sales'])
```

```python
plt.figure(figsize=(20,8))
plt.plot(sales_trend['year_month'], sales_trend['sales'], color='0.1')
plt.xticks(rotation='vertical', size=8)
plt.show()
```
